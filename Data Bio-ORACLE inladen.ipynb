#### Datasets maken met toekomstige omgevingsvariabelen van Bio-ORACLE

```python
import pandas as pd
import xarray as xr
import glob
import numpy as np
```

```python
df_biooracle = pd.read_csv("Data/Diversity_data_with_biooracle_2010.csv")
df_biooracle.columns
```

```python
df_locations = df_biooracle[["long_deg", "lat_deg"]]
df_locations.head(10)
```

#### Dataset maken voor SSP119 (2050-2060)

Alle nc bestanden tot 1 geheel samenvoegen 

```python
files56119 = glob.glob("nc2050-2060_SSP119/*.nc")
```

```python
datasets56119 = [xr.open_dataset(f) for f in files56119]
ds56119 = xr.merge(datasets56119)

print(ds56119)
```

```python
df56119 = ds56119.to_dataframe().reset_index()
```

Er zijn veel meer datapunten dan nodig

```python
len(df56119)
```

```python
df56119.head()
```

```python
df56119 = df56119.rename(columns={"longitude": "long_deg", "latitude": "lat_deg"})
```

```python
print(df56119[["lat_deg", "long_deg"]].dtypes)
df56119["lat_deg"] = df56119["lat_deg"].astype("float64").round(5)
df56119["long_deg"] = df56119["long_deg"].astype("float64").round(5)
print(df56119[["lat_deg", "long_deg"]].head())
```

Datapunten selecteren dat overeenkomen met de locaties van oorspronkelijke dataset

```python
def nearest(array, values):
    """
    array: 1D array van rasterwaarden (lat of long)
    values: array van gewenste waarden
    return: voor elke waarde in values de dichtstbijzijnde in array
    """
    array = np.array(array)
    values = np.array(values)
    idx = np.abs(array[:, None] - values).argmin(axis=0)
    return array[idx]

# vind dichtstbijzijnde lat/lon
df_locations["lat_match"] = nearest(df56119["lat_deg"].unique(), df_locations["lat_deg"])
df_locations["long_match"] = nearest(df56119["long_deg"].unique(), df_locations["long_deg"])

# merge op de “nearest” kolommen
df_subset56119 = df56119.merge(
    df_locations,
    left_on=["lat_deg", "long_deg"],
    right_on=["lat_match", "long_match"],
    how="inner"
)
```

```python
df_subset56119.head()
```

```python
df_subset56119 = df_subset56119.drop(["lat_match","long_match", "long_deg_x", "lat_deg_x", "time"], axis = 1)
df_subset56119.head()
```

```python
len(df_subset56119)
```

```python
df_subset56119 = df_subset56119.rename(columns={"long_deg_y": "long_deg", "lat_deg_y": "lat_deg"})
df_subset56119.head()
```

De onderstaande 2 blokjes code moeten True opleveren 

```python
(df_biooracle['lat_deg'] == df_subset56119['lat_deg']).all()
```

```python
(df_biooracle['long_deg'] == df_subset56119['long_deg']).all()
```

Dataset opslaan 

```python
df_subset56119.to_csv("Data/2050-60_SSP119.csv")
```
